```python
import requests
from bs4 import BeautifulSoup
url="https://github.com/"
```

```python
r=requests.get(url)
soup=BeautifulSoup(r.content,'html.parser')
```

```python
print(soup.prettify())
```

#**Scrap title of the page**

```python
title=soup.title
print(title)
```

#**Find all p tags of the page**

```python
paras = soup.find_all('p')
print(paras)
```

```python
print(soup.find('p'))
```

#**Get class of an element**

```python
print(soup.find('p')['class'])
```

```python
print(soup.find("p",class_='lead'))
```

#**Get text of an element**


```python
print(soup.find('p').get_text())
```

```python
print(soup.get_text())
```

#**Get all the anchor tags**

```python
anchors = soup.find_all('a')
print(anchors)
```

#**Get all the links on the page**

```python
all_links=set()
for link in anchors:
    if(link.get('href')!='#'):
        linkText="https://github.com/"+link.get('href')
        all_links.add(link)
        print(linkText)
```
